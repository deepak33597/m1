```
# File location and type
file_location1 = "/FileStore/tables/tableM5.csv"
file_location2 = "/FileStore/tables/tableS6.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df1 = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location1)

df2 = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location2)

temp_table_name = "table1"
df1.createOrReplaceTempView(temp_table_name)
temp_table_name2 = "table2"
df2.createOrReplaceTempView(temp_table_name2)
```

```
table2withnewtime1 = spark.sql("select a.*,concat(a.year,'-',a.month,'-',a.day,'T',a.hour,':',a.minute,':00.000+0000') newTime from (select distinct b.* from table2 b inner join (select application_id,address_type,address_line_1,address_line_2,city,state, postal_code,Country_code, validated_by_smarty_streets_indicator,creation_timestamp,max(last_updated_timestamp) last_updated_timestamp  from table2 a group by application_id,address_type,address_line_1,address_line_2,city,state, postal_code,Country_code, validated_by_smarty_streets_indicator,creation_timestamp) a on a.application_id=b.application_id and a.address_type=b.address_type and  b.last_updated_timestamp =a.last_updated_timestamp)a");
table2withnewtime1.createOrReplaceTempView("table2withnewtime");
```

```
Master_without_dublicate : having data with removed duplicate from master 
OnlyDuplicateFromMaster:having only duplicate data from master 
  
```

```
Master_without_dublicate1 = spark.sql( "select distinct a.* from table1 a inner join (select application_id,address_type ,count(*) from table1 group by application_id,address_type having count(*)=1) b on a.application_id=b.application_id and a.address_type=b.address_type");
                                      
Master_without_dublicate1.createOrReplaceTempView("Master_without_dublicate")                         
         
OnlyDuplicateFromMaster1 = spark.sql( "select a1.application_id,a1.address_type,a1.address_line_1,a1.address_line_2,a1.city,a1.state, a1.postal_code,a1.Country_code ,a1.creation_timestamp ,max(a1.last_updated_timestamp) last_updated_timestamp,a1.validated_by_smarty_streets_indicator from (select distinct a.* from table1 a inner join (select application_id,address_type,count(*) from table1 group by application_id,address_type having count(*)>1) b on a.application_id=b.application_id and a.address_type=b.address_type ) a1 group by application_id,address_type,address_line_1,address_line_2,city,state, postal_code,Country_code, validated_by_smarty_streets_indicator,creation_timestamp");                 
OnlyDuplicateFromMaster1.createOrReplaceTempView("OnlyDuplicateFromMaster")
```

```
%sql

select * from Master_without_dublicate
```

```
#taking max data from source after that we have left join with master duplicate data from master so that we can push the latest data into master duplicate
duplicate_and_source1 = spark.sql("select ifnull(a.application_id,b.application_id) application_id,ifnull(a.address_type,b.address_type) address_type,ifnull(a.address_line_1,b.address_line_1) address_line_1,ifnull(a.address_line_2,b.address_line_2) address_line_2,ifnull(a.city,b.city) city,ifnull(a.state,b.state) state,ifnull(a.postal_code,b.postal_code) postal_code,ifnull(a.Country_code,b.Country_code) Country_code,ifnull(a.creation_timestamp,b.creation_timestamp) creation_timestamp,ifnull(a.last_updated_timestamp,b.last_updated_timestamp) last_updated_timestamp, ifnull(a.validated_by_smarty_streets_indicator,b.validated_by_smarty_streets_indicator) validated_by_smarty_streets_indicator,a.newTime from OnlyDuplicateFromMaster b left join table2withnewtime a on a.application_id=b.application_id and a.address_line_1 =b.address_line_1 and a.address_type=b.address_type and ifnull(a.address_line_2,0) =ifnull(b.address_line_2,0) and a.city=b.city and a.state=b.state and a.postal_code=b.postal_code and a.Country_code=b.Country_code and b.last_updated_timestamp <=a.last_updated_timestamp")

duplicate_and_source1.createOrReplaceTempView("duplicate_and_source")
```

```
%sql
select * from duplicate_and_source;
```

```
duplicate_and_source2 = spark.sql("""select b.application_id,b.address_type,b.address_line_1,b.address_line_2,b.city,b.state,b.postal_code,b.Country_code,b.creation_timestamp,b.last_updated_timestamp,b.validated_by_smarty_streets_indicator, b.newTime from duplicate_and_source b inner join (select application_id,address_type , max(last_updated_timestamp) last_updated_timestamp from duplicate_and_source group by application_id,address_type) b1 on b.application_id=b1.application_id and b.address_type=b1.address_type and ifnull(b.last_updated_timestamp,0)=ifnull(b1.last_updated_timestamp,0)""")

duplicate_and_source2.createOrReplaceTempView("duplicate_and_source2")

duplicate_and_source3 = spark.sql("""select b.application_id,b.address_type,b.address_line_1,b.address_line_2,b.city,b.state,b.postal_code,b.Country_code,b.creation_timestamp,b.last_updated_timestamp,b.validated_by_smarty_streets_indicator from duplicate_and_source2 b inner join (select application_id,address_type , max(newTime) newTime from duplicate_and_source2 group by application_id,address_type) b1 on b.application_id=b1.application_id and b.address_type=b1.address_type and ifnull(b.newTime,0)=ifnull(b1.newTime,0)""")

duplicate_and_source3.createOrReplaceTempView("duplicate_and_source3")
```

```

duplicate_and_source_union = spark.sql("""select * from duplicate_and_source3
union all
select * from Master_without_dublicate""")

duplicate_and_source_union.createOrReplaceTempView("duplicate_and_source_union")
```

```
%sql
 select * from duplicate_and_source_union;
```

```
%sql

select * from table1 a
left join duplicate_and_source_union b
on a.application_id=b.application_id and a.address_type= b.address_type
where a.application_id is null OR a.address_type is null OR b.address_type is null OR b.application_id is null
```
