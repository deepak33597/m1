```
# File location and type
file_location1 = "/FileStore/tables/table3.csv"
file_location2 = "/FileStore/tables/table4.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df1 = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location1)

df2 = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location2)

temp_table_name = "table1"
df1.createOrReplaceTempView(temp_table_name)
temp_table_name2 = "table2"
df2.createOrReplaceTempView(temp_table_name2)
```

```
Master_without_dublicate : having data with removed duplicate from master 
OnlyDuplicateFromMaster:having only duplicate data from master 
  
```

```
Master_without_dublicate1 = spark.sql( "select distinct a.* from table1 a inner join (select application_id,address_type ,count(*) from table1 group by application_id,address_type having count(*)=1) b on a.application_id=b.application_id and a.address_type=b.address_type");
                                      
Master_without_dublicate1.createOrReplaceTempView("Master_without_dublicate")                         
         
OnlyDuplicateFromMaster1 = spark.sql( "select a1.application_id,a1.address_type,a1.address_line_1,a1.address_line_2,a1.city,a1.state, a1.postal_code,a1.Country_code ,a1.creation_timestamp ,max(a1.last_updated_timestamp) last_updated_timestamp,a1.validated_by_smarty_streets_indicator from (select distinct a.* from table1 a inner join (select application_id,address_type,count(*) from table1 group by application_id,address_type having count(*)>1) b on a.application_id=b.application_id and a.address_type=b.address_type ) a1 group by application_id,address_type,address_line_1,address_line_2,city,state, postal_code,Country_code, validated_by_smarty_streets_indicator,creation_timestamp");                 
OnlyDuplicateFromMaster1.createOrReplaceTempView("OnlyDuplicateFromMaster")
```

```
1.finding max last_updated_date  on the bais of  (application_id,address_type,address_line_1,address_line_2,city,state, postal_code,Country_code, validated_by_smarty_streets_indicator,creation_timestamp) from source 
eg source have :
  apid,typeid,state,lastupdateddate
  366,M,ca, 2019-12-25
  366,M,AR, 2019-11-24
  366,M,AR, 2019-11-24
  
  out put will be 
  apid,typeid,state,lastupdateddate
  366,M,ca, 2019-12-25
  366,M,AR, 2019-11-24 
  
2. OnlyDuplicateFromMaster will be left join  with step1 when last_updated_timestamp (source) is more then OnlyDuplicateFromMaster(master)
and then update OnlyDuplicateFromMaster.lastupdateddate from source.lastupdateddate
OnlyDuplicateFromMaster have :
  apid,typeid,state,lastupdateddate
  366,M,ca, 2019-11-30
  366,M,AR, 2019-11-30
  
  source have (from step1) :
apid,typeid,state,lastupdateddate
  366,M,ca, 2019-12-25
  366,M,AR, 2019-11-24
  
  result:
  366,M,ca, 2019-12-25
  366,M,AR, 2019-11-30
```

```

#taking max data from source after that we have left join with master duplicate data from master so that we can push the latest data into master duplicate

duplicate_and_source1 = spark.sql("select ifnull(a.application_id,b.application_id) application_id,ifnull(a.address_type,b.address_type) address_type,ifnull(a.address_line_1,b.address_line_1) address_line_1,ifnull(a.address_line_2,b.address_line_2) address_line_2,ifnull(a.city,b.city) city,ifnull(a.state,b.state) state,ifnull(a.postal_code,b.postal_code) postal_code,ifnull(a.Country_code,b.Country_code) Country_code,ifnull(a.creation_timestamp,b.creation_timestamp) creation_timestamp,ifnull(a.last_updated_timestamp,b.last_updated_timestamp) last_updated_timestamp, ifnull(a.validated_by_smarty_streets_indicator,b.validated_by_smarty_streets_indicator) validated_by_smarty_streets_indicator from OnlyDuplicateFromMaster b left join (select application_id,address_type,address_line_1,address_line_2,city,state, postal_code,Country_code, validated_by_smarty_streets_indicator ,creation_timestamp ,max(last_updated_timestamp) last_updated_timestamp  from table2 a group by application_id,address_type,address_line_1,address_line_2,city,state, postal_code,Country_code, validated_by_smarty_streets_indicator,creation_timestamp) a on a.application_id=b.application_id and a.address_line_1 =b.address_line_1 and a.address_type=b.address_type and ifnull(a.address_line_2,0) =ifnull(b.address_line_2,0) and a.city=b.city and a.state=b.state and a.postal_code=b.postal_code and a.Country_code=b.Country_code and b.last_updated_timestamp <=a.last_updated_timestamp")

duplicate_and_source1.createOrReplaceTempView("duplicate_and_source")
```

```
%sql

select * from duplicate_and_source
```

```
# duplicate_and_source2 : again inner join to remove idential data and take max timestemp

eg: we got below result  from duplicate_and_source
  apid,typeid,state,lastupdateddate
  366,M,ca, 2019-12-25
  366,M,AR, 2019-11-30
  
  Now we will take max data on the bais of  apid,typeid and will get result like
  apid,typeid,state,lastupdateddate
  366,M,ca, 2019-12-25
```

```
duplicate_and_source2 = spark.sql("""select b.application_id,b.address_type,b.address_line_1,b.address_line_2,b.city,b.state,b.postal_code,b.Country_code,b.creation_timestamp,b.last_updated_timestamp,b.validated_by_smarty_streets_indicator from duplicate_and_source b inner join (select application_id,address_type , max(last_updated_timestamp) last_updated_timestamp from duplicate_and_source group by application_id,address_type) b1 on b.application_id=b1.application_id and b.address_type=b1.address_type and b.last_updated_timestamp=b1.last_updated_timestamp""")

duplicate_and_source2.createOrReplaceTempView("duplicate_and_source2")
```

```
%sql
select * from duplicate_and_source2
```

```
Union between duplicate_and_source2 and Master_without_dublicate to get master table data with latest time stamp
```

```

duplicate_and_source_union = spark.sql("""select * from duplicate_and_source2
union all
select * from Master_without_dublicate""")

duplicate_and_source_union.createOrReplaceTempView("duplicate_and_source_union")
```

```
%sql
 select application_id,address_type, count(*) from duplicate_and_source_union where application_id in (366,424) group by application_id,address_type
```
